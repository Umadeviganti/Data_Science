### 2 c Create a function that will accept a DataFrame as input and return pie-charts for all the appropriate Categorical features. Clearly show percentage distribution in the pie-chart (4 Points)

```python
def plot_categorical_pie_charts(dataframe, columns_per_row=3):
    # Identify categorical columns
    categorical_columns = dataframe.select_dtypes(include=['object']).columns

    # Calculate the number of rows needed
    num_rows = math.ceil(len(categorical_columns) / columns_per_row)

    # Create subplots
    fig, axes = plt.subplots(num_rows, columns_per_row, figsize=(15, 5 * num_rows))
    fig.subplots_adjust(hspace=0.5)

    for i, column in enumerate(categorical_columns):
        row_idx = i // columns_per_row
        col_idx = i % columns_per_row

        # Count the occurrences of each category
        category_counts = dataframe[column].value_counts()

        # Plot a pie chart
        axes[row_idx, col_idx].pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
        axes[row_idx, col_idx].set_title(f'Percentage distribution of {column}')
    # Adjust layout
    plt.tight_layout()
    plt.show()


```

```python
#Calling a function to create a pie chart
plot_categorical_pie_charts(merged_df, columns_per_row=3)
```
