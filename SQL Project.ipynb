```python
#Trend of Customer Satisfaction
Customer_satisfaction=pd.read_csv('Customer_satisfaction.csv')
```

```python
print(Customer_satisfaction)
```

```python
# Set the quarters as the index
Customer_satisfaction.set_index('quarter_number', inplace=True)

# Create a 100% stacked bar chart
plt.figure(figsize=(10, 6))

# Define colors for each feedback category
colors = ['green', 'limegreen', 'red', 'darkred', 'orange']
```

```python
# Plot stacked bars for each feedback category
Customer_satisfaction.plot(kind='bar', stacked=True, color=colors)

plt.xlabel('Quarter Number')
plt.ylabel('Percentage')
plt.title('Trend of Customer Satisfaction')
plt.xticks(rotation=0)
plt.legend(title='Feedback Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Quarter 1 had a balanced distribution of feedback categories, with moderate percentages across all categories.

Quarter 2 showed a slight increase in negative feedback (very bad and bad) compared to Quarter 1.

Quarter 3 experienced a notable increase in very bad and bad feedback, indicating a decline in customer satisfaction.

Quarter 4 had the highest percentages of very bad and bad feedback, suggesting a significant decrease in customer satisfaction compared to the previous quarters.

Overall, there was a trend of worsening customer satisfaction over the quarters, with Quarter 4 showing the most pronounced decline. Further investigation and action may be needed to address the issues leading to this decline and improve overall customer satisfaction.

## Revenue Metrics

```python
#Trend of purchases by Quarter
Orders_by_quarter=pd.read_csv('Orders_by_quarter.csv')
```

```python
print(Orders_by_quarter)
```

```python
# Plot the trend of purchases by quarter
plt.figure(figsize=(10, 6))
plt.plot(Orders_by_quarter['quarter_number'], Orders_by_quarter['num_orders'], marker='o', color='skyblue', linestyle='-')

plt.xlabel('Quarter Number')
plt.ylabel('Number of Orders')
plt.title('Trend of Purchases by Quarter')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Overall, this trend suggests a potential decrease in demand or other factors affecting purchasing behavior as the year progresses.

```python
#Quarter on Quarter % change in Revenue
Revenue_comparision=pd.read_csv('Revenue_comparision.csv')
```

```python
print(Revenue_comparision)
```

```python
# Calculate the quarter-over-quarter percentage change in revenue
Revenue_by_quarter['QoQ % Change'] = Revenue_by_quarter['total_revenue'].pct_change() * 100
```

```python
# Plot the trend of revenue by quarter
plt.figure(figsize=(10, 6))
plt.plot(Revenue_comparision['quarter_number'], Revenue_comparision['total_revenue'], marker='o', color='skyblue', linestyle='-')

# Add quarter-over-quarter percentage change as text at each point
for i, txt in enumerate(Revenue_comparision['qoq_percentage_change']):
    plt.annotate(f'{txt:.2f}%', (Revenue_comparision['quarter_number'][i], Revenue_comparision['total_revenue'][i]))

plt.xlabel('Quarter Number')
plt.ylabel('Revenue')
plt.title('Quarter-over-Quarter % Change in Revenue')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Overall, the graph indicates a significant decline in revenue quarter-over-quarter, suggesting potential challenges or changes in the business environment that affected revenue generation over the year. Further analysis may be needed to identify the underlying factors contributing to this decline and develop strategies to address them.

```python
#Trend of Revenue and Orders by Quarter
Revenue_by_quarter=pd.read_csv('Revenue_by_quarter.csv')
```

```python
print(Revenue_by_quarter)
```

```python
# Create a figure and axis object
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot revenue on the primary y-axis
ax1.plot(Revenue_by_quarter['quarter_number'], Revenue_by_quarter['total_revenue'], color='skyblue', marker='o', label='Revenue')
ax1.set_ylabel('Revenue', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

# Create a secondary y-axis for orders
ax2 = ax1.twinx()
ax2.plot(Revenue_by_quarter['quarter_number'], Revenue_by_quarter['num_orders'], color='orange', marker='s', label='Orders')
ax2.set_ylabel('Orders', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Set labels, title, and legend
plt.xlabel('Quarter Number')
plt.title('Trend of Revenue and Orders by Quarter')
fig.legend(loc='upper left')

plt.grid(True)
plt.tight_layout()
plt.show()
```

Both revenue and the number of orders exhibit a downward trend throughout the year.

Revenue starts at a higher value in the first quarter, indicating strong initial performance, but experiences a steady decline in subsequent quarters.

Similarly, the number of orders also decreases gradually over the quarters, suggesting a reduction in customer purchasing activity.

The synchronized downward movement of revenue and orders may indicate broader market trends or internal business dynamics impacting sales performance.

## Shipping Metrics

```python
#Average discount offered by Credit Card type
Discount_on_creditcards=pd.read_csv('Discount_on_creditcards.csv')
```

```python
print(Discount_on_creditcards)
```

```python
# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(Discount_on_creditcards['credit_card_type'], Discount_on_creditcards['average_discount'], color='skyblue')
plt.xlabel('Card Types')
plt.ylabel('Average Discount')
plt.title('Average Discount Offered by Credit Card Type')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y')  # Add gridlines on the y-axis
plt.tight_layout()
plt.show()
```

The highest average discount is seen with "laser" cards, with an average discount of approximately 64.38%.

Following closely are "visa-electron" and "maestro" cards, with average discounts around 62.35% and 62.42%, respectively.

"mastercard" and "china-unionpay" also offer relatively high average discounts, around 62.95% and 62.22%, respectively.

On the lower end, "diners-club-international" and "solo" cards have the lowest average discounts, approximately 58.40% and 58.50%, respectively.
