# An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected. 

Answer the following questions. [4 Marks]

2.A. Probability that none of the items are defective? [1 Mark]

2.B. Probability that exactly one of the items is defective? [1 Mark]

2.C. Probability that two or fewer of the items are defective? [1 Mark]

2.D. Probability that three or more of the items are defective ? [1 Mark]

```python
#declare the variable for the sample i.e no of trails
n=10
```

```python
#declare the pripbability of the failure rat
p=0.05
```

```python
# Probability that none of the items are defective
prob_none_defective=binom.pmf(0,n,p)
prob_none_defective
```

```python
#Probability that exactly one of the items is defective
prob_exactly_one_defective=binom.pmf(1,n,p)
prob_exactly_one_defective
```

```python
#Probability that two or fewer of the items are defective
prob_two_or_few_defective=sum(binom.pmf(k,n,p) for k in range(3))
prob_two_or_few_defective
```

```python
# Probability that three or more of the items are defective
prob_three_or_more_defective=1-prob_two_or_few_defective
prob_three_or_more_defective
```

### Idea: 
We can keep track of how many problems (bugs) we find in the software while testing.
### Example: 
If we find 10 bugs in a program with 100 lines of code, the "bug density" is 10 bugs per 100 lines.

## Seeing When Bugs Pop Up:

### Idea: 
We can look at how bad each bug is (severity) and how urgently it needs to be fixed (priority).
### Example:
A bug that makes the whole program crash might be more severe than a bug that only affects a small part.

## How Long Does it Take to Fix Bugs:

### Idea: 
We can see on average how much time it takes to fix a reported bug.
### Example:
If, on average, it takes 3 days to fix a bug, we know what to expect for similar issues in the future.

## Finding Common Issues:

### Idea: 
We can figure out the main reasons why bugs happen and focus on fixing those issues.
### Example: 
If 80% of the bugs are caused by just 20% of the problems, we can prioritize fixing that 20%.

## Checking How Good Our Tests Are:

### Idea:

### Statistical Analysis:

#### Is there any significant difference between Funds raised by companies that are still operating vs companies that closed down?

```python

# First Filter data based on operating state
operating_data = s_data[s_data['OperatingState'] == 'Operating']
closed_data = s_data[s_data['OperatingState'] == 'Closed']

# Display descriptive statistics of seperated data
print("Operating Companies:")
print(operating_data['Funds_in_million'].describe())
print("\nClosed Companies:")
print(closed_data['Funds_in_million'].describe())
```

```python
# Now Visualization of twoseperated data set
plt.figure(figsize=(8, 6))
plt.boxplot([operating_data['Funds_in_million'], closed_data['Funds_in_million']],
            labels=['Operating', 'Closed'])
plt.title('Comparison of Funds Raised by Operating State')
plt.ylabel('Funds Raised (in million)')
plt.show()

```

As from the above we can see diffrernce between the two "Operating and closed" but as the values vary from each other, lets verify by using the hypothesis test

```python
# Hypothesis testing (t-test for independent samples)
t_stat, p_value = stats.ttest_ind(operating_data['Funds_in_million'], closed_data['Funds_in_million'])

# Display results 
print(f"\nT-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Check significance level (e.g., 0.05)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference.")
else:
    print("Fail to reject the null hypothesis. No significant difference.")
```

From the above values we can say that there is no significant difference in the funds raised between companies that are still operating and those that have closed down.  As the p-value is greater than the chosen significance level of 0.05, leading to the conclusion of failing to reject the null hypothesis.

### Write your hypothesis comparing the proportion of companies that are operating between winners and contestants:
Null Hypothesis (H0): There is no significant difference in the proportion of companies that are operating between winners and contestants.

Alternative Hypothesis (H1): There is a significant difference in the proportion of companies that are operating between winners and contestants.

```python
# Test for significance and conclusion
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(startup_data['Result'], startup_data['OperatingState'])
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

print(f'Chi-square Statistic: {chi2_stat}\nP-value: {p_value}')

# Check if p-value is less than the significance level 
if p_value < 0.05:
    print('Reject the null hypothesis. There is a significant difference in proportions.')
else:
    print('Fail to reject the null hypothesis. There is no significant difference in proportions.')
```
